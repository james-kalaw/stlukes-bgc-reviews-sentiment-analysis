###**1. NLP Procedures and Interpretation of Results**

####**A. Text Preprocessing:**

**I. Tokenization**

> Tokenization is the process of breaking down a piece of text, like a sentence or a paragraph, into individual words or “tokens.”

#Description:
The tokenized comments can be used in later stages of analysis, such as filtering stopwords, stemming, or lemmatization.

```python
reviews_df['tokenized_comments'] = reviews_df['Comments'].apply(word_tokenize)
reviews_df['tokenized_comments']

print("Tokenization")
```

**II. Stop Word Removal**

> Stop words are commonly used words that have very little meaning, such as “a,” “an,” “the,” or “in.” Stopwords are typically excluded from natural language processing (NLP) and information retrieval applications because they do not contribute much to the meaning or context of the text.

#Description:


*   Stopword Removal is a common preprocessing step in NLP to reduce noise in the data.
*   The filtered_comments column now contains only the most meaningful words from each review, which are ready for further processing or analysis.



```python
stop_words = set(stopwords.words('english'))
reviews_df['filtered_comments'] = reviews_df['tokenized_comments'].apply(lambda x: [word for word in x if word.lower() not in stop_words])
reviews_df

print("Analyzing meaning words.")
```

**III. POS Tagging and Lemmatization**

> POS Tagging involves labeling words in a text with their respective parts of speech, such as nouns, verbs, or adjectives, to understand grammatical structure and aid in text analysis. Lemmatization reduces words to their base or dictionary form (lemma) based on context and part of speech, ensuring consistent normalization; for example, "running," "ran," and "runs" are lemmatized to "run." Both techniques are essential in natural language processing for tasks like parsing and text mining.

#Description:
By lemmatizing the words, the analysis focuses on the core meaning of each word, rather than variations of the same word form.

```python
lemmatizer = WordNetLemmatizer()

def get_wordnet_pos(word):
    """Map POS tag to first character lemmatize() accepts"""
    tag = nltk.pos_tag([word])[0][1][0].upper()
    tag_dict = {"J": wordnet.ADJ,
                "N": wordnet.NOUN,
                "V": wordnet.VERB,
                "R": wordnet.ADV}

    return tag_dict.get(tag, wordnet.NOUN)

reviews_df['lemmatized_comments'] = reviews_df['filtered_comments'].apply(lambda x: [lemmatizer.lemmatize(word, get_wordnet_pos(word)) for word in x])

reviews_df['lemmatized_comments']

print("Lemmatization")
```

> Check if Lemmatization is successful

#Description:
This function checks if the lemmatization process was successful by comparing each word in the original (filtered) comments with its lemmatized counterpart.

```python
def check_lemmatization(row):
    """Check if lemmatization is successful by comparing original and lemmatized words."""
    original_words = row['filtered_comments']
    lemmatized_words = row['lemmatized_comments']

    for i in range(min(len(original_words), len(lemmatized_words))):
        if original_words[i] != lemmatized_words[i]:
            print(f"Original: {original_words[i]}, Lemmatized: {lemmatized_words[i]}")

reviews_df['lemmatized_comments'] = reviews_df['filtered_comments'].apply(lambda x: [lemmatizer.lemmatize(word, get_wordnet_pos(word)) for word in x])

if 'pos_tags' in reviews_df.columns:
    reviews_df = reviews_df.drop('pos_tags', axis=1)

reviews_df.apply(check_lemmatization, axis=1)

print("diagnossing successful lemmatization.")
```

###**2. Regression Modelling and Interpretations**



####**A. Feature Preparation:**

#Description:
Converting the text data into numerical form using TF-IDF (Term Frequency-Inverse Document Frequency) vectorization, and setting up the feature matrices for the machine learning model.

```python
x= reviews_df['lemmatized_comments']
y= reviews_df['Sentiment']

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

X_train = [' '.join(tokens) for tokens in X_train]
X_test = [' '.join(tokens) for tokens in X_test]


vectorizer = TfidfVectorizer()
X_train_tfidf = vectorizer.fit_transform(X_train)
X_test_tfidf = vectorizer.transform(X_test)

print("Feature Preparation")
```

> This script prepares text data for machine learning by splitting and transforming it. ####**A. KMeans Clustering**

#Description:
These codes perform K-Means clustering on text data, specifically on the TF-IDF features derived from the text reviews, and interprets the results by analyzing the characteristics of the clusters.

```python
# 1. Feature Preparation (Using TF-IDF features from previous steps)
# X_train_tfidf and X_test_tfidf are already available

# 2. Determine optimal number of clusters (using Elbow Method)
wcss = []  # Within-cluster sum of squares
for i in range(1, 5):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(X_train_tfidf)
    wcss.append(kmeans.inertia_)

plt.plot(range(1, 5), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()

print("Cluster Optimal Number")
```

```python
# 3. Apply KMeans with the optimal number of clusters

num_clusters = 2  # Based on Elbow Method visualization
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
clusters = kmeans.fit_predict(X_train_tfidf)

print("KMeans Clustering")
```

```python
# 4. Add cluster labels to the dataframe

reviews_df['cluster'] = np.nan

X_train_index = X_train.index if isinstance(X_train, pd.DataFrame) else list(range(len(X_train)))
reviews_df.loc[reviews_df.index.isin(X_train_index), 'cluster'] = clusters

print("Assigning Cluster")
```

```python
# 5. Analyze and Interpret Clusters

# a. Explore cluster characteristics (e.g., using word clouds)
for i in range(num_clusters):
    cluster_reviews = reviews_df[reviews_df['cluster'] == i]['lemmatized_comments']
    create_wordcloud(cluster_reviews, f'Cluster {i} Word Cloud')

print("Characteristic Clusters")
```

```python
# b. Examine cluster centroids

centroids = kmeans.cluster_centers_
feature_names = vectorizer.get_feature_names_out()

for i in range(num_clusters):
    print(f"Cluster {i}:")
    # Sort centroids by feature importance and get top 10 words
    top_words_indices = centroids[i].argsort()[-10:]
    top_words = [feature_names[j] for j in top_words_indices]
    print(top_words)
    print("\n")

print("Centroid Cluster")
```
